# Load Data

```python
import pandas as pd
import numpy as np
import os

# Open dataset
data = pd.read_csv('data/churn_data_anonymized_20200120.csv')
data.rename(columns={'survival_days': 'time', 'status': 'event'},inplace=True)

data.head()
```

# Create eval, train and test datasets and save

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Remove observations according to EDA

data = data[data['time']>=0]
data = data[data['products_created'] <= (data['products_created'].mean() + 3 * data['products_created'].std())]
data = data[data['admin_visits'] <= (data['admin_visits'].mean() + 3 * data['admin_visits'].std())]
data = data[data['tx'] <= (data['tx'].mean() + 3 * data['tx'].std())]
data = data[data['gmv_usd'] <= (data['gmv_usd'].mean() + 3 * data['gmv_usd'].std())]

data.reset_index(drop=True,inplace=True)

# Defining the features

X = pd.get_dummies(data.drop(['store_id','time', 'event'], axis=1))
T = data['time']
E = data['event']

## Create evaluation set, 70% of the data.
index_train_test, index_eval = train_test_split( range(len(data)), test_size = 0.7, random_state = 2020)

# Creating the X, T and E input
X_train_test = X.loc[index_train_test].reset_index( drop = True )
X_eval  = X.loc[index_eval].reset_index( drop = True )

T_train_test = T.loc[index_train_test].reset_index( drop = True )
T_eval  = T.loc[index_eval].reset_index( drop = True )

E_train_test = E.loc[index_train_test].reset_index( drop = True )
E_eval  = E.loc[index_eval].reset_index( drop = True )


## Create train and test set, 30% of the data.
index_train, index_test = train_test_split( range(len(X_train_test)), test_size = 0.25, random_state = 2020)

# Creating the X, T and E input
X_train = X_train_test.loc[index_train].reset_index( drop = True )
X_test  = X_train_test.loc[index_test].reset_index( drop = True )

T_train = T_train_test.loc[index_train].reset_index( drop = True )
T_test  = T_train_test.loc[index_test].reset_index( drop = True )

E_train = E_train_test.loc[index_train].reset_index( drop = True )
E_test  = E_train_test.loc[index_test].reset_index( drop = True )
```

```python
# Save the datasets.

X_train.to_csv('data/X_train_20200206.csv',index=False)
T_train.to_csv('data/T_train_20200206.csv',index=False)
E_train.to_csv('data/E_train_20200206.csv',index=False)

X_test.to_csv('data/X_test_20200206.csv',index=False)
T_test.to_csv('data/T_test_20200206.csv',index=False)
E_test.to_csv('data/E_test_20200206.csv',index=False)

X_eval.to_csv('data/X_eval_20200206.csv',index=False)
T_eval.to_csv('data/T_eval_20200206.csv',index=False)
E_eval.to_csv('data/E_eval_20200206.csv',index=False)
```
